```python
import re
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup 

startpage = int(input())
enddpage = int(input())

List = []
page = startpage
url = "https://tv.naver.com/jtbc.jtbcbros/talks"
headers = { "referer": url, "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36" } 

while True: 
    c_url = "https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?ticket=ent01&templateId=tvbrandtalk&pool=cbox6&_callback=jQuery1113014419133803426343_1598930747902&lang=ko&country=KR&objectId=3052901&categoryId=&pageSize=20&indexSize=5&groupId=&pageType=default&page="+str(page)+"&refresh=false"
    res = requests.get(c_url,headers=headers) 
    cont = BeautifulSoup(res.content, 'html.parser') 
    total_comm = str(cont).split('comment":')[1].split(",")[0] # ì‹¤ì œ ëŒ“ê¸€ split 
    
    match = re.findall('"contents":([^\*]*),"userIdNo"', str(cont)) # "content": ë’¤ì— ë‚˜ì˜¤ëŠ” ì „ì²´ ë¬¸ì¥ ëì€ ,"userIdNo"ë¡œ ëë‚˜ëŠ”...
    List.append(match) 
    
    # í•œ ë²ˆì— ëŒ“ê¸€ì´ 20ê°œì”© ë³´ì´ê¸° ë•Œë¬¸ì— í•œ í˜ì´ì§€ì”© ëª½ë•… ëŒ“ê¸€ì„ ê¸ì–´ì˜¤ê¸° 
    if int(total_comm) <= ((page*20)): 
        break 
    else: page += 1
        
    if page > enddpage:
        break
        
def flatten(I): 
    flatList = [] 
    for elem in I: 
        if type(elem) == list: #íƒ€ì…ì´ listì´ë©´ ê·¸ ì•ˆì˜ ì›ì†Œë¥¼ ì¶”ê°€ 
            for e in elem: 
                flatList.append(e)
        else: #íƒ€ì…ì´ listê°€ ì•„ë‹Œ ì›ì†Œë¼ë©´ ë°”ë¡œ ì¶”ê°€ 
            flatList.append(elem)
    return flatList 

df_naverTV_talk = pd.DataFrame(np.array(flatten(List)).T)
df_naverTV_talk.to_csv('naverTV_talk_jtbcbros_{}_{}.csv'.format(startpage, enddpage))
```
